```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
raw_df = pd.read_csv(
    "Melbourne01.csv",     
    header=None,           
    dtype=str, 
    on_bad_lines="skip",   
    skipinitialspace=True  
)

raw_df.head(10)
```

```python
raw_df.shape
raw_df.columns = [
    "Year","Month","Day","Hour","Minute","Temperature","Rainfall","DewPoint","Humidity","WindDirection","WindSpeed","WindGust","Pressure","Cloud"]

raw_df.head()
```

```python
raw_df.dtypes
```

```python
for col in raw_df.columns:
    raw_df[col] = raw_df[col].astype(str).str.strip()
raw_df.head()
```

```python
raw_df.duplicated().sum()
```

```python
raw_df = raw_df.drop_duplicates()
raw_df.shape
```

```python
raw_df.isna().sum()
```

```python
raw_df = raw_df.dropna(subset=["Month", "Day", "Hour", "Minute"])
raw_df.isna().sum()
```

```python
numeric_cols = [
    "Year",
    "Month",
    "Day",
    "Hour",
    "Minute",
    "Temperature",
    "Rainfall",
    "DewPoint",
    "Humidity",
    "WindSpeed",
    "WindGust",
    "Pressure",
    "Cloud"
]

for col in numeric_cols:
    raw_df[col] = pd.to_numeric(raw_df[col], errors="coerce")

raw_df.dtypes
```

```python
raw_df["WindDirection"].unique()
```

```python
(raw_df[["Rainfall", "WindSpeed", "WindGust", "Pressure"]] == -9999).sum()
```

```python
raw_df.loc[raw_df["Rainfall"] == -9999, "Rainfall"] = np.nan
raw_df.loc[raw_df["WindSpeed"] == -9999, "WindSpeed"] = np.nan
raw_df.loc[raw_df["WindGust"] == -9999, "WindGust"] = np.nan
raw_df.loc[raw_df["Pressure"] == -9999, "Pressure"] = np.nan

raw_df[["Rainfall", "WindSpeed", "WindGust", "Pressure"]].describe()
```

```python
raw_df = raw_df.assign(
    DateTime = pd.to_datetime(
        raw_df[["Year", "Month", "Day", "Hour", "Minute"]]
    )
)

raw_df.head()
```

```python
cleaned_df = raw_df.assign(
    WindDirection = raw_df["WindDirection"].astype("category")
)

cleaned_df.dtypes
```

```python
cleaned_df.to_csv("Melbourne01_cleaned.csv", index=False)
```
